```python
import json
from pathlib import Path
import numpy as np
import os





def flatten(list_of_list):
    return [item for sublist in list_of_list for item in sublist]

path_to_training = Path("training")
path_to_test = Path("training")  #be careful it's changed to training here


##all
training_set = ['ES2002', 'ES2005', 'ES2006', 'ES2007', 'ES2008', 'ES2009', 'ES2010', 'ES2012', 'ES2013', 'ES2015', 'ES2016', 'IS1000', 'IS1001', 'IS1002', 'IS1003', 'IS1004', 'IS1005', 'IS1006', 'IS1007', 'TS3005', 'TS3008', 'TS3009', 'TS3010', 'TS3011', 'TS3012']
training_set = flatten([[m_id+s_id for s_id in 'abcd'] for m_id in training_set])
training_set.remove('IS1002a')
training_set.remove('IS1005d')
training_set.remove('TS3012c')


to_exclude = ["<vocalsound>", "Mm-hmm", "<disfmarker>", "Uh", "<gap>", "Ach ", "Um ,", "Um", "uh", "Oh,", "Mm", " ah", "Oh ", " um "]



# Create a new folder to save modified JSON files
output_folder = "training_cleaned"
os.makedirs(output_folder, exist_ok=True)


count_modified = 0
count_total = 0

for transcription_id in training_set:
    file_path = "training" + "/" + transcription_id + '.txt'

    with open(path_to_training / f"{transcription_id}.json", "r") as file:
        transcription = json.load(file)

    X_training_text_current = []


    for utterance in transcription:
        modified = False  # Flag to check if modifications were made
        for string in to_exclude:
            if string in utterance["text"]:
                utterance["text"] = utterance["text"].replace(string, "")
                modified = True

        if modified:
            count_modified += 1
    count_total += 1

    # Save the modified transcription to a new JSON file in the output folder
    output_file_path = os.path.join(output_folder, f"{transcription_id}.json")
    with open(output_file_path, "w") as output_file:
        json.dump(transcription, output_file, indent=2)

        


print(f"\nTotal sentences processed: {count_total}")
print(f"Modified sentences: {count_modified}")


```
